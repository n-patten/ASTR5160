```python
import pymangle
import matplotlib.pyplot as plt
import healpy as hp
from numpy.random import random
import numpy as np
# NP Necessary imports
```

```python
mrect = pymangle.Mangle("/d/users/nikhil/ASTR5160/week6/rectangleclass12.ply")
# NP Creating polygon with rectangular bounds
# NP First rectangle has an area of 0.0373817088 str
# NP Task 1 complete
# NP Second rectangle has an area of 0.0385720646 str
# NP Task 2 complete
```

```python
ra = 360.*(random(1000000)) 
dec=(180/np.pi)*np.arcsin(1.-random(1000000)*2.)
# NP Generating a catalog of 1000000 points throughout sphere
# NP Task 3 complete
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
# NP Creating larger plot
plt.plot(ra, dec, 'r.', label = 'Random catalog')
# NP Plotting all randomly generated points
plt.plot(ra[mrect.contains(ra, dec)], dec[mrect.contains(ra, dec)], 'b.', label = 'Rectangular regions')
# NP Plotting all points that are within the mask
plt.title('Lat-lon rectangular polygons')
plt.xlabel('RA (deg)')
plt.ylabel('Dec (deg)')
plt.legend()
# NP Labelling plot and generating legend
plt.savefig('/d/www/nikhil/public_html/ASTR5160/Masking.png')
# NP Saving plot
# NP Taks 4 complete
```
